### Características Principales

Cada servicio web puede destacarse por las siguientes características:
* **Interoperabilidad**: Permiten que diferentes aplicaciones, desarrolladas en distintos lenguajes de programación y plataformas, se comuniquen entre sí.
* **Uso de estándares abiertos**: Utilizan protocolos y formatos estándar como HTTP, XML y JSON, lo que facilita la integración entre sistemas.

## Formatos de Intercambio de Datos 📦

Los formatos de intercambio de datos son esenciales para la comunicación entre aplicaciones. Los dos formatos más utilizados son XML y JSON. Ambos formatos tienen sus propias características y ventajas, y la elección entre uno u otro depende de las necesidades específicas de la aplicación y del entorno en el que se esté utilizando. Popularmente, JSON es preferido para aplicaciones web y móviles debido a su simplicidad y eficiencia, mientras que XML es utilizado en aplicaciones empresariales y sistemas que requieren una estructura de datos más compleja.

Según el paper ["Comparison of JSON and XML Data Formats"](https://www.researchgate.net/publication/329707959_Comparison_of_JSON_and_XML_Data_Formats) de [Dr. Alen Šimec](https://www.researchgate.net/profile/Alen-Simec), publicado en 2014, JSON no reemplazará por completo a XML, sino que ambos formatos coexistirán: JSON como la opción más práctica para flujos de datos simples y de alto rendimiento en entornos web, y XML como la herramienta de elección cuando se requiere un estándar robusto, amplio en funciones y con validación estricta de documentos.

### XML  📄

XML (eXtensible Markup Language) es un lenguaje de marcado que se utiliza para almacenar y transportar datos. Fue diseñado para ser legible tanto por humanos como por máquinas, y es ampliamente utilizado en aplicaciones web y servicios web debido a su flexibilidad y capacidad para representar datos estructurados.

### Características Principales

* **Estructura Jerárquica**: XML organiza los datos en una estructura de árbol, donde los elementos pueden contener otros elementos, creando una jerarquía clara y fácil de entender.
* **Etiquetas Personalizadas**: A diferencia de HTML, XML permite a los usuarios definir sus propias etiquetas, lo que proporciona una gran flexibilidad para representar diferentes tipos de datos.
* **Legibilidad**: XML está diseñado para ser legible tanto por humanos como por máquinas, lo que facilita la comprensión y el procesamiento de los datos.
* **Independencia de Plataforma**: XML es independiente de la plataforma y del lenguaje de programación, lo que permite su uso en una amplia variedad de aplicaciones y sistemas.
* **Validación**: XML puede ser validado utilizando DTD (Document Type Definition) o XML Schema, lo que garantiza que los datos cumplen con una estructura específica y son consistentes.
* **Soporte para Unicode**: XML soporta Unicode, lo que permite representar caracteres de diferentes idiomas y conjuntos de caracteres.

### Estructura Básica de un Documento XML

Un documento XML típico tiene la siguiente estructura básica:

```xml
    <?xml version="1.0" encoding="UTF-8"?>
    <root>
        <element attribute="value">Content</element>
        <anotherElement>
            <childElement>Child Content</childElement>
        </anotherElement>
    </root>
```

### Elementos y Atributos
* **Elementos**: Son las unidades básicas de un documento XML y están delimitados por etiquetas de apertura y cierre. Pueden contener texto, otros elementos o ambos.
* **Atributos**: Son pares clave-valor que proporcionan información adicional sobre un elemento. Los atributos se definen dentro de la etiqueta de apertura de un elemento.


### Parseo de XML con Python 🐍

El parseo de XML en Python se puede realizar utilizando diferentes bibliotecas, siendo `xml.etree.ElementTree` una de las más utilizadas debido a su simplicidad y eficiencia. A continuación, se muestra un ejemplo básico de cómo parsear un documento XML utilizando esta biblioteca.

```python
    import xml.etree.ElementTree as ET

    # Cargar y parsear el archivo XML
    tree = ET.parse('archivo.xml')
    root = tree.getroot()

    # Recorrer los elementos del XML
    for child in root:
        print(child.tag, child.attrib)
        for subchild in child:
            print('  ', subchild.tag, subchild.text)
```
En este ejemplo, se carga un archivo XML llamado `archivo.xml`, se obtiene la raíz del documento y se recorre cada elemento hijo, imprimiendo su etiqueta y atributos. Luego, se recorren los subelementos de cada hijo, imprimiendo sus etiquetas y contenido de texto.

### JSON  ☕️

JSON (JavaScript Object Notation) es un formato de intercambio de datos ligero y fácil de leer y escribir. Fue diseñado para ser simple y eficiente, y es ampliamente utilizado en aplicaciones web y servicios web debido a su compatibilidad con JavaScript y su capacidad para representar datos estructurados de manera concisa.

Es totalmente ligero y fácil de leer y escribir, lo que lo hace ideal para aplicaciones web y móviles donde la eficiencia y la velocidad son cruciales.

### Características Principales

Algunas de las características principales de JSON son:

* **Simplicidad**: Tiene una sintaxis simple y fácil de entender, lo que facilita su lectura y escritura tanto para humanos como para máquinas.
* **Estructura de Datos**: Representa datos en forma de pares clave-valor y soporta estructuras de datos complejas como objetos y arrays.
* **Compatibilidad con JavaScript**: Es nativamente compatible con JavaScript, lo que facilita su uso en aplicaciones web.
* **Independencia de Plataforma**: Es independiente de la plataforma y del lenguaje de programación, lo que permite su uso en una amplia variedad de aplicaciones y sistemas.

### Estructura

Un documento JSON típico tiene la siguiente sintaxis básica:

```json
    {
        "clave": "valor",
        "numero": 123,
        "booleano": true,
        "array": [1, 2, 3],
        "objeto": {
            "claveAnidada": "valorAnidado"
        }
    }
```

En JSON existen varios tipos de elementos:

- **Matrices (arrays):** Las matrices son listas de valores separados por comas. Las matrices se escriben entre corchetes [ ]
  ```json
    [1, "pepe", 3.14, ""]
  ```

- **Objetos (objects):** Los objetos son listas de pares nombre / valor. El nombre y el valor están separados por dos puntos : y las parejas están separadas por comas. Los objetos se escriben entre llaves { } y los nombres de las parejas se escriben siempre entre comillas dobles.

  ```json
    {
        "nombre": "Lionel Messi", 
        "edad": 38,
        "carnet de conducir": true
    }
  ```

- **Último elemento:** Tanto en los objetos como en las matrices, el último elemento no puede ir seguido de una coma.

  ```json
    {
        "nombre": "Lionel Messi",   // Esto no es válido ❌
    }
  ```

- **Sin problemas de estructura:** Los espacios en blanco y los saltos de línea no son significativos, es decir, puede haber cualquier número de espacios en blanco o saltos de línea separando cualquier elemento o símbolo del documento.

  ```json
  [
    {
      "nombre": "José Sand",
      "edad": 40,
      "carnet de conducir": true
    },
    {
      "nombre": "Jesús Dátolo",
      "edad": 38,
      "carnet de conducir": false
    }
  ]
  ```

### Parseo de JSON con Python 🐍

Parsear. ¿Qué significa parsear algo? Significa analizarlo y descomponerlo en sus partes divididas para entender su estructura y contenido. En el contexto de la programación, parsear un documento o una cadena de texto implica convertirlo en un formato que pueda ser fácilmente manipulado y procesado.

El parseo de un archivo JSON en Python se puede realizar utilizando la biblioteca estándar `json`, que proporciona funciones para convertir cadenas JSON en objetos de Python y viceversa.  Esta biblioteca ya viene incluida con Python, por lo que no es necesario instalar nada adicional.

A continuación, un ejemplo básico de cómo parsear un documento JSON utilizando esta biblioteca.

```python
    import json

    # Cargar y parsear el archivo JSON
    with open('archivo.json') as f:   # El archivo debe existir...
        data = json.load(f)

    # Acceder a los datos del JSON
    for item in data:
        print(item['clave'])
```
En este ejemplo , se carga un archivo JSON llamado `archivo.json`, se parsea su contenido y se accede a los datos del JSON, imprimiendo el valor asociado a la clave 'clave' para cada elemento en el JSON.

A simple vista pueden ser muy parecidos, pero lejos están de ser lo mismo. Por la simple razón de que:

➡️ Un Diccionario de Python es una estructura de datos nativa del lenguaje Python que almacena pares clave-valor. Aunque la representación textual de un diccionario puede parecerse a JSON, no sigue las mismas reglas estrictas de sintaxis y formato que JSON.

Decir que un JSON es un Diccionario de Python es como decir que un auto es una bicicleta. Pueden tener algunas similitudes, pero son cosas diferentes con propósitos y características distintas.



### Serialización y Deserialización

La serialización es el proceso de convertir un objeto o estructura de datos en un formato que pueda ser almacenado o transmitido, como JSON o XML. La deserialización es el proceso inverso, donde se convierte el formato almacenado o transmitido de vuelta a su forma original como un objeto o estructura de datos.

Por ejemplo:
```python
    import json

    # Serialización: Convertir un diccionario de Python a una cadena JSON
    data = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'Madrid'}
    json_string = json.dumps(data)
    print(json_string)  # Salida: {"nombre": "Juan", "edad": 30, "ciudad": "Madrid"}

    # Deserialización: Convertir una cadena JSON de vuelta a un diccionario de Python
    json_data = '{"nombre": "Ana", "edad": 25, "ciudad": "Barcelona"}'
    data_dict = json.loads(json_data)
    print(data_dict)  # Salida: {'nombre': 'Ana', 'edad': 25, 'ciudad': 'Barcelona'}
```

Aquí vemos cómo se utiliza la biblioteca `json` de Python para serializar un diccionario de Python a una cadena JSON utilizando `json.dumps()`, y luego deserializar una cadena JSON de vuelta a un diccionario de Python utilizando `json.loads()`.

El método `json.dumps()` se utiliza para escribir directamente un objeto de Python en un archivo JSON, mientras que `json.loads()` se utiliza para leer un archivo JSON y convertir su contenido en un objeto de Python.

````python
import json

# Serialización
data = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'Madrid'}  # Diccionario de Python
json_string = json.dumps(data)  # Lo convertimos a una cadena JSON
print(json_string)
print(type(json_string))  # Muestra: <class 'str'>, es una cadena JSON.
````

````python
import json

# Deserialización
json_data = '{"nombre": "Ana", "edad": 25, "ciudad": "Barcelona"}'  # Cadena JSON
data_dict = json.loads(json_data)  # Lo convertimos a un diccionario de Python
print(data_dict)
print(type(data_dict))  # Muestra: <class 'dict'>, es un diccionario de Python.
````

## API's  🛠️

Las API's (Interfaces de Programación de Aplicaciones) son un conjunto de reglas y protocolos que permiten la comunicación entre diferentes aplicaciones y servicios. Facilitan la integración de funcionalidades y datos de una aplicación en otra, lo que acelera el desarrollo y mejora la interoperabilidad.

Simplifican el desarrollo de aplicaciones y software permitiendo a los desarrolladores integrar datos, servicios y capacidades de otras aplicaciones, en lugar de desarrollarlas desde cero para cada plataforma.

Es útil pensar en la comunicación de la API en términos de una solicitud y respuesta entre un cliente y un servidor. La aplicación que envía la solicitud es el cliente y el servidor proporciona la respuesta. La API es el puente que establece la conexión entre ellos.

### REST API's

Una API REST *(Representational State Transfer)* es un conjunto de reglas y convenciones para crear servicios web que permiten la comunicación entre diferentes sistemas a través de HTTP. Las API REST son ampliamente utilizadas en aplicaciones web y móviles para acceder y manipular datos de manera eficiente.

#### Principios REST
Las API REST se basan en varios principios fundamentales:
* **Stateless (Sin Estado)**: Cada solicitud del cliente al servidor debe contener toda la información necesaria para entender y procesar la solicitud. El servidor no debe almacenar ningún estado del cliente entre solicitudes.
* **Recursos**: Los recursos son los objetos o datos que la API maneja, y cada recurso se identifica mediante una URL única. Por ejemplo, en una API de usuarios, un recurso podría ser un usuario específico identificado por su ID.
* **Métodos HTTP**: Repasando, las API REST utilizan los métodos HTTP estándar para realizar operaciones sobre los recursos. Los métodos más comunes son:
  - **GET**: Recuperar datos de un recurso.
  - **POST**: Crear un nuevo recurso.
  - **PUT**: Actualizar un recurso existente.
  - **DELETE**: Eliminar un recurso.

#### Repaso: Consumo de APIs REST desde Python (con requests)

Un repaso cortito de la Unidad 5:

```python
    import requests

    # Hacer una solicitud GET a una API REST
    response = requests.get('https://api.example.com/resource')

    # Verificar el estado de la respuesta
    if response.status_code == 200:
        data = response.json()  # Parsear la respuesta JSON
        print(data)
    else:
        print(f'Error: {response.status_code}')
```

### SOAP API's

SOAP (Simple Object Access Protocol) es un protocolo de mensajería basado en XML que se utiliza para intercambiar información estructurada en la implementación de servicios web. SOAP es independiente del lenguaje y la plataforma, lo que permite la comunicación entre aplicaciones desarrolladas en diferentes entornos. 

Aunque **REST** es hoy el estándar más popular, SOAP todavía se usa mucho en entornos empresariales y sistemas legados (por ejemplo, en banca, seguros, y gobierno).

| Característica         | REST                                  | SOAP                                  |
|-----------------------|--------------------------------------|--------------------------------------|
| **Tipo**             | Estilo arquitectónico                 | Protocolo de mensajería              |
| **Formato de datos** | JSON (principalmente) o XML           | Exclusivamente XML                   |
| **Contrato**         | No necesita contrato formal           | Requiere un WSDL (contrato)          |
| **Facilidad**        | Más simple, liviano y flexible        | Más complejo y verboso               |
| **Casos de uso**     | APIs modernas, microservicios, apps   | Sistemas legados, transacciones críticas, integración empresarial |


#### Mensajes XML y contratos WSDL

SOAP utiliza mensajes XML muy estructurados. Cada mensaje tiene un formato estándar que incluye:

- **Envelope**: indica que es un mensaje SOAP.
- **Header**: información opcional (autenticación, transacciones).
- **Body**: los datos o la operación que se quiere ejecutar.
- **Fault**: bloque de error si algo falla.

### Ejemplo de Mensaje SOAP

```xml
<soap:Envelope xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
   <soap:Header/>
   <soap:Body>
      <m:GetPrice xmlns:m="https://www.ejemplo.com/stock">
         <m:Item>Notebook</m:Item>
      </m:GetPrice>
   </soap:Body>
</soap:Envelope>
```

#### SOAP Requests con Python

Para consumir un servicio SOAP en Python, podemos usar la biblioteca `requests` que vimos en la unidad anterior. También podemos usar la librería `lxml` para construir y parsear mensajes XML.
Existen otras biblitecas especializadas como `zeep` o `suds-jurko` que facilitan mucho el trabajo con SOAP, pero aquí haremos un ejemplo simple con `requests`, dado que es la biblioteca más popular y fácil de usar.

```python
      import requests
   
      # Definir el endpoint del servicio SOAP
      url = 'https://www.ejemplo.com/soap'
   
      # Mensaje SOAP
      body = '''<?xml version="1.0"?>
      <soap:Envelope xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
         <soap:Header/>
         <soap:Body>
            <m:GetPrice xmlns:m="https://www.ejemplo.com/stock">
               <m:Item>Notebook</m:Item>
            </m:GetPrice>
         </soap:Body>
      </soap:Envelope>'''
   
      # Definir los headers HTTP
      headers = {
         'Content-Type': 'text/xml; charset=utf-8',
         'SOAPAction': 'https://www.ejemplo.com/GetPrice'
      }
   
      # Hacer la solicitud POST al servicio SOAP
      response = requests.post(url, data=body, headers=headers)
   
      # Verificar el estado de la respuesta
      if response.status_code == 200:
         print('Respuesta SOAP:')
         print(response.text)  # Aquí podríamos parsear el XML de la respuesta
      else:
         print(f'Error: {response.status_code}')
   ```

## Autenticación y Autorización 🔐

La autenticación API verifica quién realiza la solicitud; la autorización API determina a qué pueden acceder.
Los métodos de autenticación comunes incluyen autenticación básica, API key, JSON web tokens (JWT), OAuth 2.0 y otras.

### Autenticación Básica

La autenticación básica es un método simple y comúnmente utilizado para proteger las API. Funciona enviando las credenciales del usuario (nombre de usuario y contraseña) en cada solicitud HTTP. Aunque es fácil de implementar, tiene limitaciones de seguridad.

Para autenticarse, el cliente envía otra solicitud al servidor con el encabezado de autorización incluido en el encabezado de la solicitud HTTP, que contiene la palabra "Basic" seguida de una cadena codificada en base64 con el nombre de usuario y la contraseña. Por ejemplo, "Authorization: Basic dXNlcm5hbWU6cGFzc3dvcmQ=" representa las credenciales "username:password" codificadas en base64.

Si bien la autenticación básica es fácil de implementar, presenta varias limitaciones de seguridad. La más importante es que los nombres de usuario y las contraseñas se envían en texto plano, lo que los hace vulnerables a escuchas ilegales y ataques de intermediario, entre otros.

### Autenticación por API Key

La autenticación por API Key es un método comúnmente utilizado para proteger las API. Funciona mediante la asignación de una clave única (API Key) a cada usuario o aplicación que desea acceder a la API. Esta clave se envía en cada solicitud HTTP, generalmente en el encabezado de la solicitud o como un parámetro de consulta.

- Qué enviar:
La API Key se incluye en un encabezado personalizado, como 'X-API-Key', o en otro lugar definido por la API, que contiene una cadena alfanumérica única y compleja. 

- Ejemplo:
            ```json
            X-API-Key: tu_clave_api_unica_y_secreta. 
            ```

- Importante: Este método es más seguro que Basic Auth, ya que no expone el nombre de usuario y la contraseña directamente en cada solicitud, pero debes asegurar la confidencialidad de tu API Key.

### Autenticación con JSON Web Tokens (JWT)

JSON Web Tokens (JWT) es un estándar abierto (RFC 7519) que define un formato compacto y seguro para transmitir información entre partes como un objeto JSON. Los JWT son comúnmente utilizados para la autenticación y autorización en aplicaciones web y móviles debido a su eficiencia y seguridad.

Los JWT no se limitan a un solo protocolo. Pueden usarse de forma independiente en esquemas de autenticación personalizados basados ​​en tokens donde se requiere la gestión de sesiones sin estado.

Cuando el cliente envía una solicitud al servidor, incluye el JWT en un encabezado de autorización. El servidor verifica la firma del JWT mediante la clave secreta y, si la firma es válida, extrae las reclamaciones y las utiliza para autorizar la solicitud.

- Qué enviar: 
El encabezado 'Authorization' con el valor Bearer seguido del Token Web JSON. 

- Ejemplo:
            ```json
            Authorization: Bearer eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.... 
            ```

- Cómo funciona:
    - Inicio de sesión: El cliente envía credenciales (nombre de usuario/contraseña) para obtener un JWT.
    - Generación: El servidor genera el JWT con información del usuario y lo firma.
    - Verificación: El cliente incluye el JWT en las solicitudes posteriores en el encabezado Authorization.
    - Autorización: El servidor verifica la firma del JWT y autoriza la solicitud basándose en la información dentro del token, sin necesitar las credenciales originales en cada petición. 

## Contratos de Servicios 📜

Los contratos de servicios son documentos que definen las reglas, formatos y protocolos que deben seguirse para interactuar con una API o servicio web. Estos contratos son esenciales para garantizar la interoperabilidad entre diferentes sistemas y aplicaciones.

### WSDL

Un WSDL (Web Services Description Language) es un documento XML que describe un servicio web SOAP. Define las operaciones que el servicio ofrece, los mensajes que utiliza, y cómo acceder a él. Actúa como un contrato entre el proveedor del servicio y el consumidor, asegurando que ambos entiendan cómo interactuar correctamente.

Contiene:
- Las operaciones que ofrece el servicio.
- Los tipos de datos esperados (entrada y salida).
- El endpoint donde se debe llamar.
- El protocolo y el formato de los mensajes.

Ejemplo de fragmento de un WSDL:

```xml
   <wsdl:operation name="GetPrice">
      <wsdl:input message="tns:GetPriceRequest"/>
      <wsdl:output message="tns:GetPriceResponse"/>
   </wsdl:operation>
```
Piensen en un WSDL como el contrato que te dice cómo llamar al servicio.

### OpenAPI / Swagger

OpenAPI (anteriormente conocido como Swagger) es una especificación para describir y documentar APIs RESTful. Proporciona una forma estandarizada de definir los endpoints, métodos HTTP, parámetros, respuestas y otros aspectos de una API. OpenAPI facilita la comprensión y el uso de las APIs tanto para desarrolladores como para herramientas automatizadas.

Un documento OpenAPI está escrito en formato YAML o JSON y contiene información detallada sobre la API, incluyendo:
- Los endpoints disponibles y los métodos HTTP que soportan (GET, POST, PUT, DELETE, etc.).
- Los parámetros de entrada y salida.
- Los códigos de estado HTTP que puede devolver la API.
- La estructura de los datos de solicitud y respuesta (por ejemplo, usando esquemas JSON).

Ejemplo de fragmento de un documento OpenAPI en YAML:

```yaml
    openapi: 3.0.0
    info:
      title: API de Ejemplo
      version: "1.0.0"

    paths:
      /hello:
        get:
          summary: Devuelve un saludo
          responses:
            "200":
              description: Saludo exitoso
              content:
                application/json:
                  schema:
                    type: object
                    properties:
                      mensaje:
                        type: string
                        example: "¡Hola mundo!"
```
Como podemos ver, OpenAPI es una herramienta poderosa para documentar y comunicar cómo interactuar con una API RESTful.
