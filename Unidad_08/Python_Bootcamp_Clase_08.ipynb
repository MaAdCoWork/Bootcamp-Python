## Selenium 🐍

#### Antes, un poco de historia...

Durante muchos años, Mercury Interactive (hoy en día adquirida por HP) dominó el mercado de las herramientas de automatización con productos tales como QuickTest Professional (QTP) y WinRunner. Sin embargo, estas herramientas eran privadas, pesadas y con licencias costosas, lo que limitaba su accesibilidad para muchos desarrolladores y testers.

Si no pagabas, no podías trabajar.

Pero ese no era el problema principal. A inicios del año 2000, la web creció rápidamente y JavaScript comenzaba a ser un lenguaje dominante en el desarrollo web. Las aplicaciones web se volvieron más complejas y dinámicas, y QTP (Mercury) no se adaptaba bien a estos cambios. Los testers necesitaban una herramienta que pudiera manejar la naturaleza dinámica de las aplicaciones web modernas.

#### Qué sucedió entonces?

En el año 2004, el ingeniero Jason Huggins de la empresa ThoughtWorks desarrolló una herramienta interna: un pequeño script hecho en JavaScript que inyectaba código en el navegador para controlar un sitio web. Cuando sus colegas vieron el potencial de esta herramienta, decidieron compartirla con la comunidad de forma Open Source, por lo que hubo que ponerle un nombre, y allí nació Selenium.

#### Por qué "Selenium"?

El nombre "Selenium" fue elegido como una broma interna. En ese momento, Mercury Interactive estaba desarrollando un producto llamado "Mercury". Huggins y su equipo eligió el nombre "Selenium" como una referencia humorística a la idea de que su herramienta era un "antídoto" para Mercury, ya que el selenio es un elemento químico que se utiliza en pequeñas cantidades para contrarrestar los efectos tóxicos del mercurio.

Literalmente,
> _Mercury (mercurio) es tóxico._ (Automatización cerrada, cara, restrictiva).
> 
> _Selenium (selenio) se usa para contrarrestarlo._ (Open Source, gratuito, comunitario).

## Primer Script con Selenium 🐍

Una vez que ya tenemos instalado Selenium, podemos comenzar a escribir nuestro primer script para automatizar un navegador web. A continuación, te muestro un ejemplo básico de cómo abrir un navegador, navegar a una página web y cerrar el navegador.

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Abrimos el navegador Chrome
driver = webdriver.Chrome()

# Navegamos a una URL
driver.get("https://www.google.com")

# Esperamos 5 segundos
time.sleep(5)

# Cerramos el navegador
driver.quit()
````

Explicación del código:

- Se invoca a la clase `webdriver` desde el módulo `selenium`, que es la que nos permite controlar el navegador. En este caso, estamos utilizando Chrome, pero también podríamos usar Firefox, Edge u otros navegadores compatibles.

- La instanciamos en el objeto `driver`, que representa el navegador abierto.
- Usamos el método `get()` para navegar a la URL especificada (en este caso, "https://www.google.com").
- Luego hacemos una espera de 5 segundos para que podamos ver la página abierta.
- Finalmente, cerramos el navegador con el método `quit()`.

Como se puede ver, el código es bastante sencillo y directo. A medida que avancemos en la unidad, exploraremos más funcionalidades de Selenium, como la interacción con elementos web, la espera de condiciones específicas y la gestión de ventanas y alertas.

## Localizadores de Elementos Web (Selectors) 🎯

<img src="https://i.imgur.com/nGrzn1T.png">

En Selenium, los localizadores de elementos web (también conocidos como selectores) son métodos que permiten identificar y seleccionar elementos específicos en una página web para interactuar con ellos. Estos localizadores son fundamentales para automatizar tareas en navegadores web, ya que permiten a los scripts de Selenium encontrar y manipular elementos como botones, campos de texto, enlaces, etc.

Selenium proporciona varios tipos de localizadores para identificar elementos web. Los más comunes son:

1. **ID**: Selecciona un elemento por su atributo `id`. Es el localizador más rápido y confiable si el `id` es único en la página.
   ```python
   element = driver.find_element(By.ID, "id del elemento")
   ```

2. **Name**: Selecciona un elemento por su atributo `name`. Es útil cuando el `name` es único en la página.
   ```python
   element = driver.find_element(By.NAME, "nombre del elemento")
   ```
3. **Class Name**: Selecciona elementos por su atributo `class`. Puede devolver múltiples elementos si varios comparten la misma clase.
   ```python
   element = driver.find_elements(By.CLASS_NAME, "nombre de la clase")
   ```
4. **Tag Name**: Selecciona elementos por su nombre de etiqueta HTML (por ejemplo, `div`, `input`, `a`, etc.). Puede devolver múltiples elementos.
   ```python
   element = driver.find_elements(By.TAG_NAME, "nombre del tag")
   ```
5. **Link Text**: Selecciona un enlace por su texto visible completo.
   ```python
   element = driver.find_element(By.LINK_TEXT, "Link Text")
   ```
6. **Partial Link Text**: Selecciona un enlace por una parte de su texto visible.
   ```python
   element = driver.find_element(By.PARTIAL_LINK_TEXT, "Parte del Texto")
   ```
7. **CSS Selector**: Selecciona elementos utilizando selectores CSS. Es muy flexible y potente.
   ```python
   element = driver.find_element(By.CSS_SELECTOR, "css_selector")
   ```
8. **XPath**: Selecciona elementos utilizando expresiones XPath. Es muy poderoso y permite navegar por la estructura del DOM.
   ```python
   element = driver.find_element(By.XPATH, "Expresión XPath")
   ```

### Cómo elegir el Locator Adecuado

<img src="https://i.imgur.com/7Oj9xtv.png" width="600">

La elección del localizador adecuado depende de varios factores, como la estructura del HTML de la página web, la unicidad del elemento y la estabilidad del selector. Aquí hay algunas pautas para ayudarte a elegir el mejor localizador:
1. **Prioriza ID y Name**: Siempre que sea posible, utiliza `ID` o `Name`, ya que son los localizadores más rápidos y confiables debido a su unicidad en la página.
2. **Usa CSS Selectors para Flexibilidad**: Si no puedes usar `ID` o `Name`, los selectores CSS son una excelente opción debido a su flexibilidad y capacidad para combinar múltiples atributos.
3. **XPath para Estructuras Complejas**: Utiliza XPath cuando necesites navegar por estructuras HTML complejas o cuando los otros localizadores no sean suficientes.
4. **Evita Class Name y Tag Name si es Posible**: Estos localizadores pueden devolver múltiples elementos, lo que puede complicar la selección del elemento correcto.
5. **Considera la Estabilidad**: Elige localizadores que sean menos propensos a cambiar con el tiempo. Por ejemplo, los `ID` y `Name` suelen ser más estables que las clases CSS que pueden cambiar con el diseño de la página.
6. **Prueba y Verifica**: Siempre prueba tus localizadores en diferentes escenarios para asegurarte de que funcionan correctamente y seleccionan los elementos deseados.

Recuerden que dominar los localizadores de elementos web es esencial para escribir scripts de automatización efectivos y robustos con Selenium. A medida que practiques y ganes experiencia, te volverás más hábil en la selección de los localizadores adecuados para tus necesidades específicas.

También recuerden que como Automatizadores, ustedes tienen la suficiente autoridad para pedir a los desarrolladores que agreguen `ID` o `Name` únicos a los elementos que necesitan automatizar, facilitando así su trabajo. Como esto no siempre sucede en ambientes empresariales, es fundamental que dominen el uso de selectores CSS y XPath para poder manejar cualquier situación que se les presente.

## Interacciones Básicas 🤖

Hay una variedad de métodos para interactuar con los elementos web que hemos localizado utilizando los selectores mencionados anteriormente. A continuación, se describen algunas de las interacciones básicas más comunes que se pueden realizar con Selenium:

1. **Clic en un elemento**: Puedes hacer clic en un botón, enlace u otro elemento interactivo utilizando el método `click()`.
   ```python
   element.click()
   ```
2. **Enviar texto a un campo de entrada**: Puedes enviar texto a un campo de entrada (input) utilizando el método `send_keys()`.
   ```python
   element.send_keys("texto a enviar")
   ```
3. **Borrar el contenido de un campo de entrada**: Puedes borrar el contenido de un campo de entrada utilizando el método `clear()`.
   ```python
   element.clear()
   ```
4. **Obtener el texto de un elemento**: Puedes obtener el texto visible de un elemento utilizando la propiedad `text`.
   ```python
   text = element.text
   ```
5. **Obtener el valor de un atributo**: Puedes obtener el valor de un atributo específico de un elemento utilizando el método `get_attribute()`.
   ```python
    value = element.get_attribute("attribute_name")
   ```
6. **Enviar teclas especiales**: Puedes enviar teclas especiales (como Enter, Tab, etc.) utilizando la clase `Keys`.
   ```python
    from selenium.webdriver.common.keys import Keys  # Esto debe estar escrito en la línea 1 del código.
    element.send_keys(Keys.ENTER)
   ```
7. **Seleccionar opciones en un menú desplegable**: Puedes seleccionar opciones en un menú desplegable utilizando la clase `Select`.
   ```python
    from selenium.webdriver.support.ui import Select  # Esto debe estar escrito en la línea 1 del código.
    select = Select(element)
    select.select_by_visible_text("Opción")
   ```
8. **Navegar entre ventanas o pestañas**: Puedes cambiar entre diferentes ventanas o pestañas del navegador utilizando los métodos `switch_to.window()`.
   ```python
    driver.switch_to.window(window_handle)
   ```
9. **Verificar si un elemento está visible**: Puedes verificar si un elemento es visible en la página utilizando el método `is_displayed()`.
   ```python
    is_visible = element.is_displayed()
   ```
10. **Verificar si un elemento está habilitado**: Puedes verificar si un elemento está habilitado (por ejemplo, un botón) utilizando el método `is_enabled()`.
    ```python
    is_enabled = element.is_enabled()
    ```
11. **Verificar si un elemento está seleccionado**: Puedes verificar si un elemento (como una casilla de verificación o un botón de opción) está seleccionado utilizando el método `is_selected()`.
    ```python
    is_selected = element.is_selected()
    ```
12. **Manejo de iFrames**: Puedes cambiar el contexto a un iframe utilizando el método `switch_to.frame()`.
    ```python
    driver.switch_to.frame(iframe_element)
    ```
13. **Volver al contenido principal**: Puedes volver al contenido principal de la página desde un iframe utilizando el método `switch_to.default_content()`.
    ```python
    driver.switch_to.default_content()
    ```
14. **Manejo de Alertas**: Puedes aceptar o rechazar alertas utilizando los métodos `accept()` y `dismiss()`.
    ```python
    alert = driver.switch_to.alert
    alert.accept()  # Aceptar la alerta
    alert.dismiss()  # Rechazar la alerta
    texto = alert.text  # Obtener el texto de la alerta
    ```

Estas son solo algunas de las interacciones básicas que se pueden realizar con Selenium. A medida que avances en el uso de esta herramienta, descubrirás muchas más funcionalidades y métodos para interactuar con elementos web de manera más avanzada.

## Esperas y Tipos de Esperas ⏳

¿Por qué son necesarias?
Porque las webs no cargan instantáneamente. Pero no sólo eso, sino que muchos elementos dentro de una página web pueden cargarse de manera asíncrona, es decir, después de que la página principal ya se ha cargado. Esto puede causar problemas en los scripts de automatización si intentan interactuar con elementos que aún no están disponibles en el DOM (Document Object Model).

Entonces, ¿usamos el método `time.sleep()` para pausar la ejecución del script?
No. No es recomendable. Si bien es algo que podría funcionar, también puede hacer que los scripts sean ineficientes y propensos a fallos, ya que no garantiza que el elemento esté disponible después de un tiempo fijo. En su lugar, Selenium proporciona mecanismos de espera más sofisticados y flexibles.

Entonces, ¿qué son las esperas en Selenium?
Las esperas son mecanismos que permiten pausar la ejecución del script hasta que se cumpla una condición específica, como la presencia de un elemento en la página web o la carga completa de la página. Son esenciales para garantizar que los scripts de automatización funcionen de manera confiable, especialmente cuando se trabaja con aplicaciones web dinámicas donde los elementos pueden tardar en cargarse o cambiar de estado.

Existen dos tipos principales de esperas en Selenium: esperas explícitas y esperas implícitas.

### Esperas Explícitas ⏳

Selenium permite esperar hasta que algo ocurra, no esperar "porque sí".

Las esperas explícitas permiten definir una condición específica que debe cumplirse antes de continuar con la ejecución del script. Se utilizan junto con la clase `WebDriverWait` y las condiciones esperadas (`expected_conditions`) para esperar hasta que un elemento esté presente, visible, clickeable, etc. Aquí hay un ejemplo de cómo usar una espera explícita para esperar a que un elemento sea clickeable:

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Crear una instancia de WebDriverWait con un tiempo máximo de espera de 10 segundos
wait = WebDriverWait(driver, 10)

# Esperar hasta que el elemento sea clickeable
element = wait.until(EC.element_to_be_clickable((By.ID, "element_id")))

# También podemos optar por esperar hasta que el elemento esté localizado
element = wait.until(EC.presence_of_element_located((By.ID, "element_id")))

# Hacer clic en el elemento una vez que sea clickeable.
# A este punto, ya sabemos que el elemento está listo para ser interactuado.
element.click()
```

### Esperas Implícitas ⏳
Las esperas implícitas establecen un tiempo de espera predeterminado para que el WebDriver busque elementos en la página. Si un elemento no está presente de inmediato, el WebDriver esperará hasta que el tiempo especificado haya transcurrido antes de lanzar una excepción. Aquí hay un ejemplo de cómo configurar una espera implícita:

```python
# Establecer una espera implícita de 10 segundos
driver.implicitly_wait(10)

# Ahora, al buscar elementos, el WebDriver esperará hasta 10 segundos si no los encuentra de inmediato
element = driver.find_element(By.ID, "element_id")
```

## Manejo de Errores 🚨

La automatización NO debe romperse ante el primer problema. Debe fallar bien, explicar por qué y limpiar los recursos utilizados (cerrar el navegador, archivos abiertos, conexiones, etc.).
Para manejar errores en Selenium, podemos utilizar bloques `try-except` para capturar excepciones específicas que puedan ocurrir durante la ejecución del script. Nosotros ya vimos en unidades anteriores cómo funcionan los bloques `try-except` en Python, así que ahora veremos cómo aplicarlos en el contexto de Selenium:

```python
from selenium.common.exceptions import NoSuchElementException

try:
    element = driver.find_element(By.ID, "non_existent_id")
    element.click()
except NoSuchElementException:
    print("El elemento no fue encontrado en la página.")
finally:
    driver.quit()
```

Como pueden ver, se captura el error `NoSuchElementException`, que ocurre cuando Selenium no puede encontrar un elemento en la página. En el bloque `except`, podemos manejar el error de manera adecuada, como imprimir un mensaje o tomar alguna acción correctiva. El bloque `finally` asegura que el navegador se cierre correctamente, independientemente de si ocurrió un error o no.

Nunca dejamos un error "crudo" sin manejar.

> Recuerden siempre: Un fallo de automatización no es un crash, es información.

## Manejo de Timeouts ⏳

En Selenium, los timeouts son períodos de tiempo que se establecen para esperar a que ocurran ciertas condiciones antes de que se lance una excepción. Son esenciales para garantizar que los scripts de automatización funcionen de manera confiable, especialmente cuando se trabaja con aplicaciones web dinámicas donde los elementos pueden tardar en cargarse o cambiar de estado.

```Python
from selenium.common.exceptions import TimeoutException

try:
    # Crear una instancia de WebDriverWait con un tiempo máximo de espera de 10 segundos
    wait = WebDriverWait(driver, 10)

    # Esperar hasta que el elemento sea clickeable
    element = wait.until(EC.element_to_be_clickable((By.ID, "element_id")))

    # Hacer clic en el elemento una vez que sea clickeable.
    element.click()
except TimeoutException:
    print("El tiempo de espera para encontrar el elemento ha expirado.")
finally:
    driver.quit()
```
