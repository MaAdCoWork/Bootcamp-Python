## Pytest 🦆

Pytest es un framework de testing para Python que facilita la escritura y ejecución de pruebas automatizadas. Es ampliamente utilizado en la comunidad de desarrollo debido a su simplicidad, flexibilidad y capacidad para manejar pruebas complejas.
Pero, ¿qué significa eso? Imaginate que estás construyendo un edificio. Necesitás asegurarte de que cada parte (los cimientos, las paredes, las ventanas) esté bien hecha, ¿verdad? Bueno, Pytest te ayuda a hacer eso con tu código. Te permite escribir "tests" que verifican que tu código haga lo que se supone que tiene que hacer.

### ¿Por qué Pytest y no otra cosa? 🤷‍♀️

Hay varias razones por las que Pytest es tan popular:

* **Es fácil de usar:** La sintaxis es sencilla y no requiere mucho boilerplate (código repetitivo).
* **Es flexible:** Se adapta a diferentes tipos de proyectos, desde pequeños scripts hasta grandes aplicaciones.
* **Tiene un montón de plugins:** Hay plugins para casi todo lo que se te ocurra, desde testing de APIs hasta cobertura de código.
* **Es compatible con otros frameworks:** Si ya estás usando otro framework de testing, como `unittest`, podés integrarlo con Pytest. (Mi recomendación: quedate con Pytest).

Es especialmente útil para equipos de desarrollo y QA que desean analizar y compartir los resultados de sus pruebas de manera efectiva.

Te da una forma visual de ver qué pruebas pasaron, cuáles fallaron, y por qué. Es mucho más fácil de entender que solo mirar una lista de resultados en la consola.

* **Soporte para adjuntos:** Permite agregar capturas de pantalla, logs y otros archivos a los informes.

## Y cómo funcionan juntos? 🤔

Cuando combinás Pytest con Allure Reports y Selenium, obtenés una poderosa herramienta para automatizar pruebas web y generar informes detallados. Pytest se encarga de ejecutar las pruebas, Selenium interactúa con la aplicación web, y Allure Reports genera informes visuales de los resultados.

Acá hay un flujo básico de cómo funcionan juntos:
1. **Escribís pruebas con Pytest:** Usás Pytest para definir tus pruebas automatizadas. Estas pruebas pueden incluir interacciones con una aplicación web utilizando Selenium.
2. **Ejecutás las pruebas:** Cuando ejecutás tus pruebas con Pytest, este framework maneja la ejecución y registra los resultados.
3. **Selenium interactúa con la web:** Durante la ejecución de las pruebas, Selenium se encarga de abrir el navegador, navegar por la aplicación web, y realizar las acciones definidas en tus pruebas (como hacer clic en botones, completar formularios, etc.).
4. **Generás informes con Allure Reports:** Después de que las pruebas se ejecutan, Allure Reports toma los resultados registrados por Pytest y genera informes visuales detallados. Estos informes incluyen información sobre qué pruebas pasaron, cuáles fallaron, capturas de pantalla, logs, y más.
5. **Analizás los resultados:** Finalmente, podés revisar los informes generados por Allure Reports para entender el estado de tu aplicación web y detectar cualquier problema que necesite ser abordado.

### Qué pasa si sólo automatizás Backend?

Si solo estás automatizando pruebas de backend, el flujo es similar, pero sin la parte de Selenium. En este caso, Pytest se encarga de ejecutar las pruebas que interactúan directamente con la lógica del backend (como APIs, bases de datos, etc.), y Allure Reports genera los informes basados en esos resultados. Aunque no hay interacción con una interfaz web, los beneficios de tener informes visuales y detallados siguen siendo muy valiosos para analizar el estado de tu backend.

## Buenas Prácticas de Uso ✅

   - Organiza tus archivos de prueba en una estructura lógica. Por ejemplo, podés tener una carpeta `tests/` con subcarpetas para diferentes módulos o funcionalidades.

## Aplicando Pytest 🧪

En esta sección vamos a dar el primer paso formal en el uso de Pytest como framework de testing.
El objetivo no es solo "hacer que un test pase", sino entender cómo Pytest descubre, ejecuta y valida tests.

### Estructura de Carpetas

Una estructura clara es fundamental para mantener el proyecto escalable y ordenado.

**Puntos importantes:**
- Todos los tests deben vivir dentro de la carpeta `test/`.
- Pytest buscará automáticamente tests dentro de este directorio.
- Todos los tests deberán llevar el prefijo `test_` o el sufijo `_test.py`. Por qué? Pytest detecta los tests por convención de nombres. No por configuración.
- Le agregamos la numeración la inicio para ordenarlos mejor visualmente.
- Además, dentro de los tests, las funciones también deben comenzar con `test_`.  Si no, Pytest no las reconocerá como tests.
- No mezclar lógica de negocio con tests.

## Asserts ✔️

Las assertions son fundamentales en cualquier framework de testing, y Pytest no es la excepción. Una assertion es una declaración que verifica si una condición es verdadera. Si la condición es falsa, la prueba falla.

Qué diferencia hay entre usar `assert` y otras formas de validación? Bueno, `assert` es una forma simple y directa de validar condiciones en Python. Pytest mejora la experiencia de las assertions al proporcionar mensajes de error detallados automáticamente cuando una assertion falla. Si ejecutas un assert mediante Pytest, la consola te mostrará exactamente qué salió mal, lo que facilita la depuración. En cambio, si usás otras formas de validación, como `if` statements con excepciones personalizadas, tenés que manejar los mensajes de error vos mismo, lo que puede ser más tedioso y menos informativo.

Algunos ejemplos comunes de asserts:

````python
import pytest

assert 5 > 3, "5 debería ser mayor que 3."
assert "python" in "python bootcamp", "La palabra 'python' debería estar en 'python bootcamp'."
assert len([1, 2, 3]) == 3, "La longitud de la lista debería ser 3."
````

Comparaciones:

````python
import pytest

actual = True
esperado = True
assert actual == esperado, "El valor actual debería ser igual al esperado."
esperado = False
assert actual != esperado, "El valor actual debería ser diferente al esperado."
assert actual > 0, "El valor actual debería ser mayor que cero."
````

Como vemos acá, Pytest muestra mensajes claros cuando una aserción falla, lo que facilita la identificación del problema.

El error es legible y el error ya no es un mensaje "crudo" de Python, sino un mensaje claro que indica qué condición no se cumplió.

Es importante destacar que Pytest ofrece una amplia gama de funcionalidades adicionales para trabajar con assertions, como la capacidad de agrupar múltiples assertions, manejar excepciones esperadas y más. Esto hace que Pytest sea una herramienta poderosa y flexible para la validación de pruebas en Python.

Por qué actual es mayor que 0? Porque en Python, los valores booleanos `True` y `False` se comportan como los enteros `1` y `0`, respectivamente. Entonces, cuando hacemos la comparación `actual > 0`, estamos verificando si `True` (que es equivalente a `1`) es mayor que `0`, lo cual es cierto.

Veamos un ejemplo donde un assert falla:

````python
import pytest

def test_fallido():
    actual = True
    esperado = False
    assert actual == esperado, "El valor actual debería ser igual al esperado."

test_fallido()
````

Aquí se ve claramente cómo Pytest informa el fallo de la aserción, mostrando el mensaje personalizado que definimos.

### Escribiendo el Primer Test

Vamos a crear un test simple para validar una función básica. Primero, creamos un archivo llamado `01_login_test.py` dentro de la carpeta `test/`.  La ruta sería así:

```bash
test/01_sumar_test.py
```

Dentro de este archivo, escribimos el siguiente código:

```python
import pytest

def test_sumar():
    resultado = 2 + 2
    assert resultado == 4, "La suma de 2 + 2 debería ser 4."
```

Este test ejecuta una suma, y utiliza una aserción para verificar que el resultado sea correcto. Si la aserción falla, Pytest marcará el test como fallido, y mostrará el mensaje de error proporcionado.

### Cómo ejecutar los Tests

Para ejecutar los tests que hemos escrito con Pytest, simplemente abrí una terminal o consola, navegá hasta el directorio raíz de tu proyecto y ejecutá el siguiente comando:

```bash
pytest
```

Esto buscará automáticamente todos los archivos y funciones de prueba en el directorio `test/` y ejecutará las pruebas encontradas. Verás un resumen de los resultados en la consola, indicando cuántas pruebas pasaron, fallaron o fueron omitidas.

Si sólamente se busca ejecutar un archivo de test específico, podés hacerlo proporcionando la ruta al archivo:

```bash
pytest test/01_sumar_test.py
```

Esto ejecutará únicamente las pruebas definidas en ese archivo.

En cambio, si querés ejecutar una función de prueba específica dentro de un archivo, podés usar la siguiente sintaxis:

```bash
pytest test/01_sumar_test.py::test_sumar
```

Esto ejecutará únicamente la función `test_sumar` dentro del archivo `01_sumar_test.py`.

## Marcadores 🏷️

Los marcadores permiten clasificar, agrupar y ejecutar selectivamente tests.
Son fundamentales en proyectos reales donde no todos los tests se ejecutan siempre.

En automatización profesional **no se corre todo siempre**.

Un marcador es una etiqueta que se le asigna a un test para indicar:

- Qué tipo de test es
- Cuándo debería ejecutarse
- A qué grupo pertenece

Ejemplos típicos:
- smoke
- regression
- critical
- slow
- ui
- api
o bien, el nombre de una funcionalidad específica del test.
- login
- consulta_saldos
- etc.

### Formas de Ejecución

Ejemplo simple:

```python
import pytest

@pytest.mark.login
def test_login_exitoso():
    assert True
```
Aquí, hemos marcado el test `test_login_exitoso` con el marcador `login`. Es una forma de categorizar este test como relacionado con la funcionalidad de login. Como si "lo bautizáramos".

Entonces, al momento de ejecutar tests usando marcadores, podemos seleccionar cuáles correr.  Por ejemplo, para ejecutar solo los tests marcados con `login`, usaríamos:

```bash
pytest -m "login"    (puede ir con o sin comillas)
```

Ejecutar tests que no tengan un marcador específico:

```bash
pytest -m "not login"
```

Ejecutar múltiples marcadores:

```bash
pytest -m "login or smoke"
```

Ejecutar tests que tengan ambos marcadores:

```bash
pytest -m "login and smoke"
```

### Registro de Marcadores

Para evitar advertencias sobre marcadores desconocidos, es recomendable registrar los marcadores personalizados en el archivo `pytest.ini` en la raíz del proyecto. Esto ayuda a documentar los marcadores que estás utilizando y evita mensajes de advertencia durante la ejecución de las pruebas.

```ini
[pytest]
markers =
    login: Tests relacionados con la funcionalidad de login.
    smoke: Tests de humo para validación rápida.
    regression: Tests de regresión para funcionalidades existentes.
```

### Marcadores con Allure

Allure Reports también soporta el uso de marcadores para categorizar y organizar los tests en los informes generados. Al utilizar marcadores en tus pruebas, podés aprovechar las funcionalidades de Allure para filtrar y visualizar los resultados de manera más efectiva.

Por ejemplo, si marcás un test con `@pytest.mark.login`, Allure lo reconocerá y podrás filtrar los resultados en el informe para ver solo los tests relacionados con el login. Esto es especialmente útil cuando tenés un gran conjunto de pruebas y querés enfocarte en una categoría específica.

```Python
import allure
import pytest

@allure.title("Login exitoso")
@pytest.mark.login
def test_login_ok():
    assert True
```
Acá marcamos el test `test_login_ok` con el marcador `login` y también le dimos un título descriptivo utilizando el decorador `@allure.title`. Cuando generes el informe de Allure, podrás filtrar los resultados para ver solo los tests relacionados con el login, lo que facilita la navegación y el análisis de los resultados.

Para integrar Pytest con Allure Reports, necesitás seguir algunos pasos para asegurarte de que ambos funcionen juntos correctamente.

    Después de ejecutar las pruebas, podés generar el informe de Allure utilizando el siguiente comando:

- **Reporte HTML Estático:**


    Luego, abrí el archivo `index.html` dentro del directorio `allure-report` en tu navegador para ver el informe.

### Allure Steps

Allure Steps es una característica de Allure Reports que permite dividir las pruebas en pasos lógicos y documentar cada uno de ellos. Esto ayuda a mejorar la legibilidad de los informes y facilita la identificación de problemas específicos en el flujo de la prueba.

Los steps (pasos) permiten que el reporte muestre el test como una "historia".  En vez de tener un solo bloque de información, podés ver cada paso que se ejecutó, lo que facilita entender qué hizo el test y dónde pudo haber fallado. Es decir:

- Paso 1: Abrir login.
- Paso 2: Ingresar credenciales.
- Paso 3: Validar Productos.

Ejemplo Básico:

```python
import pytest
import allure

def test_login_ok():
    with allure.step("Abrir SauceDemo"):
        # Código para abrir la página de login
        pass

    with allure.step("Ingresar credenciales"):
        # Código para ingresar usuario y contraseña
        pass

    with allure.step("Validar Productos"):
        # Código para validar que se muestran los productos
        assert True
```

## Screenshots en Allure 📸

El screenshot es la evidencia Nº1 en UI testing.
La idea es tomar capturas de pantalla en momentos clave de la ejecución del test, como cuando ocurre un error o al finalizar una acción importante. Estas capturas se adjuntan automáticamente al reporte de Allure, proporcionando una referencia visual que ayuda a entender el estado de la aplicación en esos momentos específicos.

Podemos definir una función para tomar screenshots y adjuntarlos al reporte de Allure:

```python
import allure

def get_screenshot(driver, name="screenshot"):
    png = driver.get_screenshot_as_png()
    allure.attach(png, name=name, attachment_type=allure.attachment_type.PNG)
```
Luego, podemos llamar a esta función en nuestros tests en los momentos que consideremos importantes:

```python
def test_example(driver):
    driver.get("https://www.saucedemo.com/")
    get_screenshot(driver, "Home")
    assert "Swag Labs" in driver.title
```

## Integrando Selenium con Pytest 🔌

En la Unidad 8 aprendimos Selenium como “automatización de navegador”.
Ahora, el objetivo es que Selenium deje de ser scripts sueltos y pase a ser parte de una **suite de tests**:

- repetible.
- mantenible.
- ejecutable por lotes.
- con setup/teardown consistente.

### Configurar Selenium como fixture

La integración correcta empieza por una regla:

> **El test NO debe crear el driver.**  
> El driver debe proveerse por fixture.

Esto evita duplicación, mejora el orden del proyecto, y hace que todos los tests arranquen bajo condiciones consistentes.

### Ejemplo base (driver por test)

Archivo recomendado: `tests/conftest.py`

**¿Qué es conftest.py?**

Es un archivo especial en Pytest que permite compartir fixtures y configuraciones entre múltiples archivos de test. Pytest lo detecta automáticamente y lo usa para cargar fixtures globales. Se debe ubicar en la raíz del proyecto o en la carpeta de tests.

```python
import pytest
from selenium import webdriver

@pytest.fixture
def driver():
    driver = webdriver.Chrome()
    driver.maximize_window()
    yield driver
    driver.quit()
```

**¿Qué logra esto?**
- Cada test recibe un navegador nuevo.
- Al terminar el test, se cierra siempre.
- El test se enfoca en validar, no en preparar el entorno.

**Ejemplo de uso en un test:**

Precondición: Tener el archivo `tests/conftest.py` con el fixture `driver` definido.

```python
import pytest

def test_open_saucedemo(driver):
    driver.get("https://www.saucedemo.com/")
    assert "Swag Labs" in driver.title
    print("Página cargada correctamente.")
```

> Pytest inyecta automáticamente el fixture `driver` en el test `test_open_saucedemo`.

Al ejecutar este test, Pytest utilizará el fixture `driver` definido en `conftest.py`, creando una instancia de Chrome, maximizando la ventana, y cerrándola al finalizar el test. Esto asegura que cada test tenga un entorno limpio y consistente para ejecutar las pruebas de Selenium.

## Fixtures

Los **fixtures** son uno de los conceptos más importantes de Pytest.  
Permiten preparar y limpiar el contexto de ejecución de los tests (setup/teardown) de manera **reutilizable, limpia y profesional**.

Son funciones especiales que se utilizan para configurar y limpiar el entorno de prueba antes y después de la ejecución de los tests. Son ideales para manejar tareas repetitivas como la inicialización de objetos, la configuración de bases de datos, o en nuestro caso, la gestión del navegador con Selenium.

En automatización con Selenium, un fixture suele ser el lugar ideal para:
- crear y configurar el `driver`
- inicializar sesión (login)
- preparar datos
- limpiar estado (logout, cerrar navegador)
- capturar evidencias ante fallas

Un fixture es una función decorada con `@pytest.fixture` que Pytest ejecuta **antes** del test para proveer un recurso, y opcionalmente ejecuta limpieza **después** del test.

> En vez de repetir “setup/teardown” en cada test, lo centralizamos en fixtures.

#### Ejemplo simple (sin Selenium)

```python
import pytest

@pytest.fixture
def numeros():
    return [1, 2, 3]

def test_longitud_lista(numeros):
    assert len(numeros) == 3
```

### Alcance de los Fixtures

Por defecto, los fixtures tienen un alcance de función, es decir, se ejecutan antes y después de cada test que los utiliza. Sin embargo, Pytest permite definir otros alcances para los fixtures, lo que puede ser útil en diferentes escenarios.

El scope define cuántas veces Pytest ejecuta el fixture:

| Scope | Se ejecuta... | Uso típico |
|-------|----------------|------------|
| function | antes y después de cada test que lo usa | máximo aislamiento |
| class | una vez por clase | suite agrupada |
| module | una vez por módulo (archivo) | tests relacionados |
| session | una vez por sesión de Pytest (corrida) | configuración global |

Ejemplo:

```python
import pytest

@pytest.fixture(scope="function")
def datos():
    return {"env": "test"}
```

`function` es el valor por defecto y es el más seguro: cada test arranca "limpio".


### Fixtures reutilizables

Un fixture reutilizable debe:

- Tener una sola responsabilidad.
- Ser fácil de entender.
- Servir para muchos tests.
- Evitar hardcodear lógica específica de un test.

**Antipatrón (malo): fixture con demasiadas responsabilidades**

```python
import pytest

@pytest.fixture
def driver_y_login_y_carrito():
    ...
```

Este fixture hace demasiado: crea el driver, hace login, y prepara un carrito de compras. Si un test solo necesita el driver, está forzado a cargar toda esa lógica innecesaria.

**Buen patrón (bueno): fixture con responsabilidad única**

```python
import pytest

@pytest.fixture
def driver():
    ...

@pytest.fixture
def login(driver):
    ...
```

Cada fixture tiene una responsabilidad clara y puede ser combinado según las necesidades de cada test.


### Fixtures para setup/teardown

En automatización, es normal necesitar acciones repetidas antes y después de cada test, como abrir y cerrar un navegador, loguearse, volver a un estado base, etc. Los fixtures nos permiten hacer esto de manera ordenada y reutilizable. Y en Pytest, los fixtures son fáciles de definir y usar para estos casos.

### Fixture de Espera Explícita

```python
import pytest
from selenium.webdriver.support.ui import WebDriverWait

@pytest.fixture
def wait(driver):
    return WebDriverWait(driver, 10)
```

Este fixture `wait` proporciona una instancia de `WebDriverWait` que puede ser utilizada en los tests para implementar esperas explícitas, mejorando la estabilidad de las pruebas al esperar condiciones específicas antes de continuar con las acciones.

### Fixture para ir a un Login

```python
import pytest

@pytest.fixture
def go_to_login(driver):
    driver.get("https://www.saucedemo.com/")
    return driver
```

Este fixture `go_to_login` navega a la página de login de la aplicación antes de que se ejecute cualquier test que lo utilice. Esto asegura que todos los tests que dependan de la página de login comiencen desde el mismo punto.

### Fixture de Login Ok

```python
import pytest

@pytest.fixture
def login_ok(driver):
    driver.get("https://www.saucedemo.com/")
    driver.find_element("id", "user-name").send_keys("standard_user")
    driver.find_element("id", "password").send_keys("secret_sauce")
    driver.find_element("id", "login-button").click()

    # Validación mínima de que entró
    wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "title")))

    return driver
```

Este fixture `login_ok` realiza un login exitoso en la aplicación. Navega a la página de login, completa las credenciales y hace clic en el botón de login. Además, incluye una validación mínima para asegurarse de que el login fue exitoso antes de devolver el driver para su uso en los tests.

### Modo de uso en tests

```python
def test_login_page_title(go_to_login):
    assert "Swag Labs" in go_to_login.title

def test_login_success(login_ok):
    assert "Products" in login_ok.title
```

Aquí, los tests `test_login_page_title` y `test_login_success` utilizan los fixtures `go_to_login` y `login_ok`, respectivamente. Esto permite que cada test comience en el estado adecuado sin necesidad de repetir el código de configuración dentro de cada test.

Esto ya va tomando forma de un proyecto real.

## Page Object Model (POM) 📦

<img src="https://i.imgur.com/5WD1bXB.jpeg" width="600">

El **Page Object Model (POM)** es un patrón de diseño que propone una regla simple:

> **Cada página de la aplicación se representa como una clase.**  
> Esa clase encapsula la forma de interactuar con la página.

El archivo de test **no debería conocer**:
- selectores.
- detalles del HTML.
- cómo se hace click.
- cómo se escribe en inputs.

Eso es responsabilidad del *Page Object*.

Recondando, la estructura era:

```bash
pages/
 ├── login_page.py
 └── products_page.py

tests/
 └── test_login.py
```

Los selectores y la lógica de interacción con la página van en el Page Object.  Los tests sólo llaman a métodos del Page Object, y en los test solamente habrán asserts y lógica de validación.

### Page Object de ejemplo: Login Page

Archivo `pages/po_login.py`

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

class PoLogin:

    def __init__(self, driver, wait):
        self.driver = driver
        self.wait = wait

    def open(self, url):
        self.driver.get(url)

    def login(self, username, password):
        self.wait.until(EC.visibility_of_element_located((By.ID, "user-name"))).send_keys(username)
        self.driver.find_element(By.ID, "password").send_keys(password)
        self.driver.find_element(By.ID, "login-button").click()
```

Archivo `tests/test_login.py`

```python
def test_login_ok(driver, wait):
    login_page = LoginPage(driver, wait)
    login_page.open("https://www.saucedemo.com/")
    login_page.login("standard_user", "secret_sauce")

    assert "inventory" in driver.current_url
```

Como podemos ver, el test `test_login_ok` no conoce los selectores ni la forma de interactuar con la página de login. Simplemente utiliza el Page Object `LoginPage` para abrir la página y realizar el login. Esto hace que los tests sean más limpios, fáciles de leer y mantener.
En cambio, la clase `PoLogin` encapsula toda la lógica de interacción con la página de login, incluyendo los selectores y las acciones necesarias para completar el proceso de login.
