## Programación Orientada a Objetos (POO) 🐍

<center><img src="https://i.imgur.com/29p9mgQ.jpeg" width="600"></center>

La Programación Orientada a Objetos (POO) es un paradigma de programación que organiza el código en torno a objetos y clases, en lugar de centrarse únicamente en funciones o procedimientos.

En Python, casi todo es un objeto (números, cadenas, listas, funciones, etc.), lo que hace que este paradigma sea muy natural de aprender y aplicar.

* Busca representar entidades del mundo real como "objetos" en el código (Un auto, una persona, un banco, etc).
* Cada objeto tiene atributos (características) y métodos (comportamientos).
* Facilita la reutilización del código y la organización.
* La unidad fundamental de la POO es la **clase**, que es una plantilla para crear objetos.

### Programación Estructurada vs POO

Hay muchas diferencias entre la programación estructurada y la programación orientada a objetos. A continuación, se presentan algunas de las diferencias clave:

| Programación Estructurada               | Programación Orientada a Objetos (POO) |
|----------------------------------------|-----------------------------------------|
| Se centra en funciones y procedimientos| Se centra en objetos y clases           |
| Datos y funciones están separados      | Datos y funciones están encapsulados en objetos |
| Menos reutilización de código          | Mayor reutilización de código           |
| Más difícil de mantener proyectos grandes | Más fácil de mantener proyectos grandes |
| Ejemplo: C                             | Ejemplo: Python, Java, C++              |

Como puedes ver, la POO ofrece varias ventajas sobre la programación estructurada, especialmente en términos de organización, reutilización y mantenimiento del código.

## Objetos y Clases 🏗

En la Programación Orientada a Objetos (POO), los conceptos de "objeto" y "clase" son fundamentales. Los conceptos de 'Atributos' y 'Métodos' también son esenciales para entender cómo funcionan los objetos y las clases en POO, pero los veremos más adelante.

### ¿Qué es una Clase?

Una clase es una plantilla o un molde que define las características y comportamientos comunes de un conjunto de objetos. Es como un plano que describe cómo se debe construir un objeto. Si buscamos una analogía, una clase es como un plano arquitectónico para construir casas, o bien una receta para preparar un plato.

Ejemplo:

> Se define la Clase "Auto" que tiene ciertas características (atributos) como color, marca y modelo, y ciertos comportamientos (métodos) como arrancar, acelerar y frenar.  El nombre de la clase será "Auto", definida con mayúscula inicial, sin espacios y en singular.

> Se define la Clase "Animal" que tiene ciertas características (atributos) como nombre y edad, además de ciertos comportamientos (métodos) como comer, dormir, etc.


### ¿Qué es un Objeto?

Un objeto es una instancia concreta de una clase. Es decir, cuando creas un objeto a partir de una clase, estás creando una entidad específica que tiene las características y comportamientos definidos por esa clase. Siguiendo con la analogía anterior, un objeto sería una casa construida siguiendo el plano arquitectónico o un plato preparado siguiendo la receta.

Ejemplo:

> A partir de la clase "Auto", se crean objetos específicos como un auto rojo de marca Toyota y modelo Corolla, o un auto azul de marca Ford y modelo Focus.
> Cada uno de estos autos es un objeto distinto, pero ambos comparten las características y comportamientos definidos en la clase "Auto".
> El nombre del objeto será en minúscula, sin espacios y en singular, por ejemplo "auto", "un_auto", "mi_auto", "auto_toyota", "auto_ford", etc.

> Otro ejemplo: A partir de la clase "Animal", se crean objetos específicos como un perro llamado "Pancho" de 4 años, o un gato llamado "Paco" de 3 años.

En POO, todo puede ser considerado un objeto. Pero si abstraemos un poco más el concepto, cada objeto puede pertenecer a una clase que define su tipo, sus caracteristicas y comportamientos.

## Atributos

Los atributos son las características o propiedades que describen un objeto. En la Programación Orientada a Objetos (POO), los atributos se definen dentro de una clase y se utilizan para almacenar información sobre el estado del objeto.
Por ejemplo, en una clase `Auto`, los atributos podrían incluir `color`, `marca` y `modelo`. Estos atributos permiten que cada instancia (objeto) de la clase `Auto` tenga sus propias características únicas.

```python
class Auto:
    def __init__(self, color, marca, modelo):
        self.color = color
        self.marca = marca
        self.modelo = modelo
```

En este ejemplo, `color`, `marca` y `modelo` son atributos de la clase `Auto`. Cuando se crea un objeto de esta clase, estos atributos se inicializan con los valores proporcionados al constructor `__init__`.

¿No es a caso lo que habíamos puesto antes? Exactamente, los atributos son las variables que definen las características del objeto, y se valen de un constructor para ser inicializados automáticamente al momento de crear una instancia de la clase.

### Atributos de Instancia vs Atributos de Clase

Los atributos pueden clasificarse en dos tipos principales: atributos de instancia y atributos de clase. Ambos tipos de atributos tienen diferentes propósitos y comportamientos.

1. **Atributos de Instancia**: Son aquellos que pertenecen a una instancia específica de una clase. Cada objeto creado a partir de la clase tiene su propia copia de estos atributos. Se definen dentro del constructor utilizando `self`. Por ejemplo:

```python
    class Auto:
        def __init__(self, color, marca, modelo):
            self.color = color
            self.marca = marca
            self.modelo = modelo
```

En este caso, `color`, `marca` y `modelo` son atributos de instancia. Cada objeto `Auto` tendrá sus propios valores para estos atributos.

2. **Atributos de Clase**: Son aquellos que pertenecen a la clase en sí, y no a una instancia específica. Se definen directamente dentro de la clase, pero fuera de cualquier método. Todos los objetos de la clase comparten la misma copia de estos atributos. Por ejemplo:

```python
    class Auto:
        cantidad_ruedas = 4  # Atributo de clase

        def __init__(self, color, marca, modelo):
            self.color = color
            self.marca = marca
            self.modelo = modelo
```

Aquí, `cantidad_ruedas` es un atributo de clase. Todos los objetos `Auto` compartirán el mismo valor para `cantidad_ruedas`, que es 4.

>Los atributos de instancia son específicos de cada objeto, mientras que los atributos de clase son compartidos por todos los objetos de la clase.

## Métodos

Los métodos son funciones definidas dentro de una clase que describen los comportamientos o acciones que un objeto puede realizar. En la Programación Orientada a Objetos (POO), los métodos permiten que los objetos interactúen con sus propios datos (atributos) y realicen operaciones específicas.

Por ejemplo, en una clase `Auto`, podríamos tener métodos como `arrancar`, `acelerar` y `frenar`. Estos métodos definen las acciones que un objeto de la clase `Auto` puede llevar a cabo.

```python
class Auto:
    def __init__(self, color, marca, modelo):
        self.color = color
        self.marca = marca
        self.modelo = modelo

    def arrancar(self):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está arrancando.")

    def acelerar(self, velocidad):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está acelerando a {velocidad} km/h.")

    def frenar(self):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está frenando.")
```

En este ejemplo, `arrancar`, `acelerar` y `frenar` son métodos de la clase `Auto`. Cada método puede acceder a los atributos del objeto utilizando `self` y realizar acciones específicas relacionadas con el objeto.

Si no usas `self`, los métodos no podrán acceder a los atributos del objeto, lo que limitará su funcionalidad y capacidad para interactuar con los datos del objeto.

### Métodos de Instancia vs Métodos de Clase

Los métodos en la Programación Orientada a Objetos (POO) pueden clasificarse en dos tipos principales: métodos de instancia y métodos de clase. Ambos tipos de métodos tienen diferentes propósitos y comportamientos.

1. **Métodos de Instancia**: Son aquellos que operan sobre una instancia específica de una clase. Estos métodos pueden acceder y modificar los atributos del objeto utilizando `self`. Se definen con `self` como el primer parámetro. Por ejemplo:

```python
class Auto:
    def __init__(self, color, marca, modelo):
        self.color = color
        self.marca = marca
        self.modelo = modelo

    def arrancar(self):
        print(f"El {self.color} '{self.marca} {self.modelo}' está arrancando.")
```

En este caso, `arrancar` es un método de instancia que puede acceder a los atributos del objeto.

2. **Métodos de Clase**: Son aquellos que operan sobre la clase en sí, en lugar de una instancia específica. Estos métodos no pueden acceder a los atributos de instancia directamente, pero pueden acceder a los atributos de clase utilizando `cls`. Se definen con `@classmethod` y `cls` como el primer parámetro. Por ejemplo:

```python
class Auto:
    cantidad_ruedas = 4  # Atributo de clase

    def __init__(self, color, marca, modelo):
        self.color = color
        self.marca = marca
        self.modelo = modelo

    @classmethod
    def mostrar_cantidad_ruedas(cls):
        print(f"Todos los autos tienen {cls.cantidad_ruedas} ruedas.")
```

Aquí, `mostrar_cantidad_ruedas` es un método de clase que accede al atributo de clase `cantidad_ruedas`.  Esto significa que no necesita una instancia específica de `Auto` para ser llamado; se puede llamar directamente en la clase.

>Entonces, los métodos de instancia operan sobre objetos específicos, mientras que los métodos de clase operan sobre la clase en sí.

### Métodos Estáticos

Los métodos estáticos son un tipo especial de método en la Programación Orientada a Objetos (POO) que no operan ni sobre una instancia específica de una clase ni sobre la clase en sí. En lugar de eso, los métodos estáticos son funciones que pertenecen a una clase, pero no tienen acceso a los atributos o métodos de instancia o de clase.

Se definen utilizando el decorador `@staticmethod` y no requieren `self` o `cls` como primer parámetro. Los métodos estáticos son útiles cuando se desea agrupar funciones relacionadas con una clase, pero que no necesitan acceder a los datos de la clase o de sus instancias.

Ejemplo de un método estático en una clase `Auto`:

```python
class Auto:
    def __init__(self, color, marca, modelo):
        self.color = color
        self.marca = marca
        self.modelo = modelo

    @staticmethod
    def es_auto_valido(marca):
        marcas_validas = ["Toyota", "Ford", "Honda", "Chevrolet"]
        return marca in marcas_validas
```

En este ejemplo, `es_auto_valido` es un método estático que verifica si una marca de auto es válida. No necesita acceder a los atributos del objeto o de la clase, por lo que se define como un método estático.

## Definición de Clases en Python

En Python, las clases se definen utilizando la palabra clave `class`, seguida del nombre de la clase y dos puntos. Dentro de la clase, se pueden definir atributos (variables) y métodos (funciones) que describen el comportamiento del objeto.
Sigamos con el ejemplo de la clase "Auto".

```python

class Auto:
    def arrancar(self):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está arrancando.")
 
    def acelerar(self, velocidad):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está acelerando a {velocidad} km/h.")

    def frenar(self):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está frenando.")
```

En este ejemplo, se define una clase llamada `Auto` con tres métodos: `arrancar`, `acelerar` y `frenar`. Cada método utiliza `self` para acceder a los atributos del objeto (como `color`, `marca` y `modelo`), que se supone que se han definido en algún lugar del objeto al que hace referencia.

Acá tenemos 2 cosas nuevas:  el indicador `self` y algo que nos hace falta para que el código funcione correctamente: un constructor para inicializar los atributos del objeto.  ¿Qué son estas cosas? Ahí vamos.

### El rol de Self

En Python, dentro de una clase, el primer parámetro de los métodos es siempre `self`. Este parámetro representa la instancia actual del objeto y permite acceder a sus atributos y métodos desde dentro de la clase.

Dicho en palabras fáciles, `self` es una referencia al objeto que está siendo manipulado. Cuando se llama a un método en un objeto, Python automáticamente pasa el objeto como el primer argumento al método, y ese argumento se denomina `self`.

### Constructores

Primero empecemos definiendo qué es un constructor en el contexto de la Programación Orientada a Objetos (POO).  Un constructor es un método especial que se utiliza para inicializar un objeto cuando se crea una instancia de una clase.

Entonces, ¿qué hace un constructor técnicamente? Establece los valores iniciales de los atributos del objeto. Esto es importante porque asegura que el objeto esté en un estado válido desde el momento en que se crea.

Sin el constructor, tendríamos que establecer manualmente los atributos del objeto después de crearlo, lo que podría llevar a errores o inconsistencias si olvidamos hacerlo.

En Python, el constructor es un método especial llamado `__init__` y su uso es fundamental para la creación de objetos. Cada vez que se crea un objeto, el constructor `__init__` se llama automáticamente para inicializar el objeto con los valores proporcionados.

Ejemplo de cómo se define un constructor en una clase `Auto`:

```python
class Auto:   # primero se define la clase
    def __init__(self, color, marca, modelo):  # este es el constructor
        self.color = color    # Mediante el uso de self se asignan los atributos
        self.marca = marca
        self.modelo = modelo
```

Como podrán ver, el constructor `__init__` toma varios parámetros (color, marca, modelo) que se utilizan para inicializar los atributos del objeto cuando se cree. El uso de `self` permite que cada instancia del objeto tenga sus propios valores para estos atributos.

## Creación de una clase Básica y su instancia en Objetos

Vamos a ver un ejemplo sencillo de cómo definir una clase llamada `Auto` y cómo crear un objeto (instancia) que provenga de esa clase.

````python
class Auto:
    def __init__(self, color, marca, modelo):
        self.color = color
        self.marca = marca
        self.modelo = modelo

    def arrancar(self):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está arrancando.")
        
    def acelerar(self, velocidad):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está acelerando a {velocidad} km/h.")

    def frenar(self):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está frenando.")

# Crear una instancia (objeto) de la clase Auto, proporcionando sus Atributos
mi_auto = Auto("rojo", "Toyota", "Corolla")

# Llamar a sus métodos
mi_auto.arrancar()
mi_auto.acelerar(velocidad=80)
mi_auto.frenar()
````

### Buenas Prácticas para Organizar Clases, Objetos, Métodos y Atributos

Al organizar métodos y atributos dentro de una clase en Python, es importante seguir ciertas buenas prácticas para mejorar la legibilidad, mantenibilidad y coherencia del código. Aquí hay algunas recomendaciones:

1. **Orden de los Métodos y Atributos**:
     - Escribir los atributos de clase al principio de la clase.
     - Seguir con el constructor `__init__`.
     - Luego, organizar los métodos en el siguiente orden:
     - Métodos de instancia
     - Métodos de clase  (en el caso que hubiere).
     - Métodos estáticos  (en el caso que hubiere).
     - Siempre es recomendable crear un diagrama de clases para tener una visión clara de la estructura del sistema y sus relaciones.

2. **Nombres Claros y Descriptivos**:
    - Utilizar nombres claros y descriptivos para los atributos y métodos.
    - Evitar abreviaciones innecesarias que puedan dificultar la comprensión del código.
    - Seguir las convenciones de nomenclatura de Python (por ejemplo, usar `snake_case` para métodos y atributos).

3. **Documentación**:
    - Incluir docstrings para la clase y sus métodos para explicar su propósito y uso.
    - Esto ayuda a otros desarrolladores (y a ti mismo en el futuro) a entender rápidamente qué hace cada parte del código.

## Abstracción

La abstracción es un concepto fundamental en la Programación Orientada a Objetos (POO) que se refiere al proceso de simplificar y representar entidades complejas mediante la identificación de sus características esenciales y la omisión de detalles innecesarios. En otras palabras, la abstracción permite centrarse en los aspectos relevantes de un objeto o sistema, mientras se ignoran los detalles que no son importantes para el contexto actual.

En pocas palabras:
> La abstracción simplifica la complejidad mostrando qué hace un objeto, sin obligar a conocer cómo lo hace.

La abstracción permite:
* Enfocarse en el comportamiento y las características esenciales de un objeto.
* Simplificar la interacción con objetos complejos.
* Crear modelos más manejables y comprensibles.
* Facilitar la reutilización del código al centrarse en interfaces y comportamientos comunes.
* Reducir el acoplamiento entre componentes del sistema.

### Por qué es necesaria la Abstracción

Imaginemos un auto real: para conducirlo solo necesitamos volante, acelerador, frenos, etc.
No necesitamos saber qué ocurre internamente en el motor o en la computadora de a bordo para manejarlo.

En la Programación Orientada a Objetos pasa lo mismo:
Queremos que los objetos expongan solo la parte útil, no sus detalles internos.

### Clases Abstractas

<img src="https://i.imgur.com/T7mAg8V.png" width="400">

Las clases abstractas son un concepto avanzado en la Programación Orientada a Objetos (POO) que permite definir una estructura común para un conjunto de clases relacionadas, sin proporcionar una implementación completa. Una clase abstracta no puede ser instanciada directamente; en su lugar, sirve como una plantilla para que otras clases hereden de ella y proporcionen implementaciones específicas para los métodos definidos como abstractos.

Siguiendo el ejemplo anterior, es sólo para abstraer aún más el concepto de clases.
Supongamos, si tengo un objeto "ferrari" y otro objeto llamado "kawasaki", lo primero que se nos viene a la mente es pensar en 2 clases distintas ("Auto" y "Moto"), pero si pensamos un poco más, ambos objetos comparten ciertas características y comportamientos comunes, como "color", "marca", "modelo", "arrancar", "acelerar", etc.  Entonces, ¿por qué no abstraer aún más el concepto y crear una **clase abstracta** llamada "Vehículo" que defina estas características y comportamientos comunes?

Es así como nacen las clases abstractas, que permiten definir una estructura común para un conjunto de clases relacionadas, sin proporcionar una implementación completa.

De este modo, las clases "Auto" y "Moto" pueden heredar atributos y métodos de la clase abstracta "Vehículo", y cada una puede implementar sus propios detalles específicos. Por ejemplo, la clase "Auto" podría tener un método específico para abrir las puertas, mientras que la clase "Moto" podría tener un método para hacer wheelie.



### Interfaces (Simulación en Python)

En la Programación Orientada a Objetos (POO), una interfaz es un contrato que define un conjunto de métodos que una clase debe implementar, sin proporcionar detalles sobre cómo se implementan esos métodos. Las interfaces permiten definir comportamientos comunes para diferentes clases, asegurando que todas las clases que implementen la interfaz tengan los mismos métodos con las mismas firmas.

Asique, pasando en limpio, una interfaz es una clase más, sólo que no tiene atributos ni implementaciones de métodos, sólo la definición de los métodos que deben ser implementados por las clases que la hereden, es decir, sus firmas/nombres y parámetros que recibe.

Ahora bien, existen las interfaces en Python? No! Python no tiene un concepto formal de interfaces como otros lenguajes de programación (por ejemplo, Java o C#). Sin embargo, se pueden simular interfaces utilizando clases abstractas del módulo `abc` (Abstract Base Classes, forma parte de la biblioteca estándar de Python) o simplemente definiendo clases con métodos que deben ser implementados por las subclases.

Aquí hay un ejemplo de cómo simular una interfaz en Python utilizando clases abstractas:

````python
from abc import ABC, abstractmethod

class Vehiculo(ABC):
    # Aquí sólo estarán las firmas de métodos SIN DESARROLLAR.
    @abstractmethod
    def arrancar(self):
        pass

    @abstractmethod
    def acelerar(self, velocidad):
        pass
    
    @abstractmethod
    def frenar(self):
        pass
    
class Auto(Vehiculo):
    def arrancar(self):
        print("El auto está arrancando.")
    def acelerar(self, velocidad):
        print(f"El auto está acelerando a {velocidad} km/h.")
    def frenar(self):
        print("El auto está frenando.")

class Moto(Vehiculo):
    def arrancar(self):
        print("La moto está arrancando.")
    def acelerar(self, velocidad):
        print(f"La moto está acelerando a {velocidad} km/h.")
    def frenar(self):
        print("La moto está frenando.")
        
# Creando instancias de Auto y Moto
auto = Auto()
moto = Moto()
auto.arrancar()
auto.acelerar(100)
moto.arrancar()
moto.acelerar(80)
````

## Encapsulamiento

El encapsulamiento es un principio fundamental de la Programación Orientada a Objetos (POO) que se refiere a la práctica de restringir el acceso directo a los datos y comportamientos internos de un objeto. En otras palabras, el encapsulamiento implica ocultar los detalles internos de un objeto y exponer solo una interfaz pública para interactuar con él.

Esto ayuda a proteger la integridad del objeto, evitando que su estado interno sea modificado de manera inapropiada o inesperada desde fuera del objeto. Al encapsular los datos y comportamientos, se promueve la modularidad, la reutilización del código y la facilidad de mantenimiento.

Ejemplo Básico de Encapsulamiento en Python:

```python
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        self.__saldo = saldo_inicial  # Atributo privado

    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto
            print(f"Depósito de {monto} realizado. Nuevo saldo: {self.__saldo}")
        else:
            print("El monto a depositar debe ser positivo.")

    def retirar(self, monto):
        if 0 < monto <= self.__saldo:
            self.__saldo -= monto
            print(f"Retiro de {monto} realizado. Nuevo saldo: {self.__saldo}")
        else:
            print("Fondos insuficientes o monto inválido.")

    def obtener_saldo(self):
        return self.__saldo
```

En este ejemplo, la clase `CuentaBancaria` encapsula el atributo `__saldo`, que es privado y no puede ser accedido directamente desde fuera de la clase. En su lugar, se proporcionan métodos públicos como `depositar`, `retirar` y `obtener_saldo` para interactuar con el saldo de manera controlada.

En resumen:
* El encapsulamiento protege los datos internos de un objeto.
* Promueve la modularidad y la reutilización del código.
* Facilita el mantenimiento al permitir cambios internos sin afectar la interfaz pública.
* Mejora la seguridad al evitar modificaciones inapropiadas del estado del objeto.
* La mejor forma de acceder a un atributo privado es mediante un método público que controle su acceso. (Ejemplo, `obtener_saldo()` en el ejemplo anterior).

### Niveles de Acceso

En Python, los niveles de acceso a los atributos y métodos de una clase se pueden controlar mediante convenciones de nomenclatura. Aunque Python no tiene modificadores de acceso explícitos como otros lenguajes (por ejemplo, `public`, `private`, `protected`), se utilizan ciertas convenciones para indicar el nivel de acceso deseado.

1. **Atributos y Métodos Públicos**:
   - Son accesibles desde cualquier parte del código, tanto dentro como fuera de la clase.
   - No tienen ningún prefijo especial en su nombre.
   - Ejemplo:
     ```python
     class MiClase:
         def metodo_publico(self):
             print("Este es un método público.")
     ```
2. **Atributos y Métodos Protegidos**:
   - Se indican con un solo guion bajo (`_`) al inicio del nombre.
   - Sugieren que son de uso interno y no deben ser accedidos directamente desde fuera de la clase o sus subclases.
   - Ejemplo:
     ```python
     class MiClase:
           def _metodo_protegido(self):
                print("Este es un método protegido.")
     ```
3. **Atributos y Métodos Privados**:
   - Se indican con dos guiones bajos (`__`) al inicio del nombre.
   - No son accesibles directamente desde fuera de la clase. Python realiza un proceso llamado "name mangling" para evitar el acceso directo.
   - Ejemplo:
     ```python
     class MiClase:
         def __metodo_privado(self):
             print("Este es un método privado.")
     ```

En resumen, aunque Python no tiene modificadores de acceso estrictos, las convenciones de nomenclatura ayudan a indicar la intención del desarrollador sobre cómo deben ser utilizados los atributos y métodos de una clase.

## Herencia

La herencia en la Programación Orientada a Objetos (POO) permite crear nuevas clases basadas en clases existentes. La clase existente se denomina "clase base" o "superclase", "clase padre", mientras que la nueva clase se llama "clase derivada" o "subclase", también "clase hija". La subclase hereda los atributos y métodos de la superclase, lo que permite reutilizar el código y extender la funcionalidad de la clase base.

Con herencia podemos crear jerarquías de clases, donde las subclases pueden especializarse y agregar comportamientos específicos, mientras mantienen la estructura y funcionalidad común definida en la superclase.  Si recuerdan el ejemplo anterior de abstracción de clases, la clase abstracta "Vehículo" podría ser la superclase, y las clases "Auto" y "Moto" serían subclases que heredan de "Vehículo". Sin darnos cuenta estamos aplicando el concepto de herencia ahí mismo, dado que ambas clases comparten atributos y métodos comunes definidos en la clase "Vehículo".

Ejemplo de Herencia en Python:

```python
class Vehiculo:
    def __init__(self, color, marca):
        self.color = color
        self.marca = marca

    def arrancar(self):
        print(f"El {self.color} '{self.marca}' está arrancando.")

class Auto(Vehiculo):
    def __init__(self, color, marca, modelo):
        super().__init__(color, marca)  # Llamada al constructor de la superclase
        self.modelo = modelo

    def acelerar(self, velocidad):
        print(f"El auto {self.color} '{self.marca} {self.modelo}' está acelerando a {velocidad} km/h.")

class Moto(Vehiculo):
    def __init__(self, color, marca, tipo):
        super().__init__(color, marca)  # Llamada al constructor de la superclase
        self.tipo = tipo

    def hacer_wheelie(self):
        print(f"La moto {self.color} '{self.marca} {self.tipo}' está haciendo un wheelie.")
```

En este ejemplo, la clase `Vehiculo` es la superclase que define atributos y métodos comunes. Las clases `Auto` y `Moto` son subclases que heredan de `Vehiculo` y agregan sus propios atributos y métodos específicos.



### Reutilización de Código en Clases Hijas

La herencia permite que las clases hijas reutilicen el código de las clases padres, lo que facilita la creación de nuevas clases sin tener que duplicar el código existente. Al heredar de una clase padre, una clase hija puede acceder a los atributos y métodos definidos en la clase padre, lo que promueve la reutilización y la consistencia en el diseño del código.

Ejemplo de Reutilización de Código en Clases Hijas:

````python
class Animal:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def comer(self):
        print(f"{self.nombre} está comiendo.")
        
class Perro(Animal):
    def ladrar(self):
        print(f"{self.nombre} está ladrando.")
        
class Gato(Animal):
    def maullar(self):
        print(f"{self.nombre} está maullando.")
        
# Creación de objetos
mi_perro = Perro("Rex", 3)
mi_gato = Gato("Miau", 2)

# Uso de métodos heredados y específicos
mi_perro.comer()  # heredado
mi_perro.ladrar()  # Método específico de Perro
mi_gato.comer()  # heredado
mi_gato.maullar()  # Método específico de Gato
````

## Herencia Múltiple

Este es uno de los conceptos más avanzados de la Programación Orientada a Objetos (POO). La herencia múltiple es una característica que permite a una clase heredar atributos y métodos de más de una clase padre o superclase. Esto significa que una subclase puede combinar comportamientos y características de múltiples clases, lo que puede ser útil para modelar relaciones complejas entre objetos.

En Java esto no es posible. Sólo se permite heredar de una única clase padre. En Python, en cambio, la herencia múltiple es totalmente válida y se puede aplicar en ciertas situaciones donde sea necesario. No es algo que conceptualmente esté mal, pero hay que usarla con precaución para evitar complicaciones en el diseño del código.

### Ventajas y Riesgos de la Herencia Múltiple

La herencia múltiple ofrece varias ventajas, pero también conlleva ciertos riesgos y desafíos que deben ser considerados cuidadosamente al diseñar sistemas orientados a objetos. A continuación, se presentan algunas de las ventajas y riesgos asociados con la herencia múltiple:

| Ventajas | Riesgos |
|----------|---------|
| **Reutilización de Código**: Permite reutilizar código de múltiples clases, lo que puede reducir la duplicación y mejorar la mantenibilidad. | **Complejidad**: Puede aumentar la complejidad del diseño del sistema, lo que dificulta la comprensión y el mantenimiento del código. |
| **Flexibilidad**: Facilita la creación de clases que combinan comportamientos de diferentes clases, lo que puede ser útil para modelar relaciones complejas. | **Conflictos de Nombres**: Puede surgir el problema del "diamante" cuando dos superclases tienen métodos o atributos con el mismo nombre, lo que puede llevar a ambigüedades sobre cuál método se debe utilizar. |
| **Modularidad**: Permite una mayor modularidad al permitir que las clases se centren en comportamientos específicos y luego se combinen según sea necesario. | **Dificultad en la Depuración**: La depuración puede volverse más complicada debido a la interacción entre múltiples clases y la posible ambigüedad en la resolución de métodos. |

En resumen, la herencia múltiple puede ser una herramienta poderosa en la POO, pero debe utilizarse con precaución y consideración para evitar problemas de diseño y mantenimiento en el código. Es importante evaluar cuidadosamente si la herencia múltiple es la mejor solución para un problema específico antes de implementarla en un sistema.

### Ejemplo de Herencia Múltiple en Python

Aquí veremos un ejemplo en Python de herencia múltiple contextualizado en Personas y sus diferentes idiomas.

````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        return f"Hola, soy {self.nombre} y tengo {self.edad} años."

class HablanteDeIngles:
    def hablar_ingles(self):
        return "Hello, how are you?"

class HablanteDeEspanol:
    def hablar_espanol(self):
        return "Hola, ¿cómo estás?"

class Bilingue(Persona, HablanteDeIngles, HablanteDeEspanol):
    def presentarse_bilingue(self):
        return (
            f"{self.presentarse()} "
            "Puedo hablar inglés y español."
        )

persona = Bilingue("Gonzalo", 30)

print(persona.presentarse_bilingue())
print(persona.hablar_ingles())
print(persona.hablar_espanol())
````

## Poliformismo

El **polimorfismo** es uno de los pilares fundamentales de la Programación Orientada a Objetos (POO).  
La palabra viene del griego: *poly* (muchos) y *morphé* (forma), y significa **“muchas formas”**.

En POO, el polimorfismo permite que **un mismo método pueda comportarse de forma distinta según el objeto que lo invoque**, incluso aunque los objetos pertenezcan a clases diferentes.

En términos simples:

> Distintas clases pueden implementar **métodos con el mismo nombre**, pero con comportamientos diferentes.  
> El código que llama a esos métodos no necesita saber de qué clase exacta es cada objeto.

Gracias al polimorfismo se logra:

- Mayor **flexibilidad** en el diseño.
- Código más **genérico y reutilizable**.
- Menos **acoplamiento** entre componentes.
- Posibilidad de **extender** el sistema sin modificar código ya existente (solo agregando nuevas clases que respeten la misma interfaz de métodos).

Python soporta polimorfismo de manera muy natural.  
En Python, el foco no está tanto en el **tipo estático** del objeto, sino en **qué sabe hacer** ese objeto (qué métodos tiene).

Hay dos enfoques típicos:

1. **Polimorfismo por herencia**  
   Distintas subclases heredan de una clase base y **sobrescriben** (override) algunos métodos.

2. **Polimorfismo por duck typing**  
   Clases completamente independientes pueden ser usadas de forma intercambiable si implementan los mismos métodos.


### Polimorfismo por Herencia (Method Overriding)

En python, la sobreescritura de métodos (method overriding) es una característica de la Programación Orientada a Objetos (POO) que permite a una subclase proporcionar una implementación específica de un método que ya está definido en su superclase. Esto significa que la subclase puede modificar el comportamiento del método heredado para adaptarlo a sus necesidades específicas.

Ejemplo de Sobreescritura de Métodos en Python:



````python
class Animal:
    def hacer_sonido(self):
        print("El animal hace un sonido.")
        
class Perro(Animal):
    def hacer_sonido(self):
        print("El perro ladra.")
        
class Gato(Animal):
    def hacer_sonido(self):
        print("El gato maúlla.")
        
# Crear instancias de las clases
mi_perro = Perro()
mi_gato = Gato()

# Llamar al método hacer_sonido en cada instancia
mi_perro.hacer_sonido()  # Salida: El perro ladra.
mi_gato.hacer_sonido()   # Salida: El gato maúlla.
````

### Poliformismo por Duck Typing

Python lleva el polimorfismo más allá de la herencia mediante el concepto de duck typing.

La idea es:

> “Si camina como un pato, nada como un pato y hace _cuak_, entonces es un pato”.

Es decir, lo que importa no es el tipo concreto de la clase, sino si el objeto tiene los métodos apropiados.

No necesitamos que todas las clases hereden de una misma superclase; basta con que implementen el método que vamos a usar.

Ejemplo de duck typing: distintos “seres” que pueden hablar()

````python
class Perro:
    def hablar(self):
        return "Guau!"

class Gato:
    def hablar(self):
        return "Miau!"

class Robot:
    def hablar(self):
        return "01001000 01101111 01101100 01100001"  # "Hola" en binario


# Duck Typing
seres = [Perro(), Gato(), Robot()]

for ser in seres:
    # No importa qué tipo de objeto es.
    # Solo nos importa que tenga un método hablar().
    print(ser.hablar())
````

### En resumen...

El polimorfismo permite que distintas clases respondan al mismo mensaje (método) con comportamientos propios.

En Python, el polimorfismo se usa:
- A través de herencia (subclases que sobrescriben métodos).
- A través de duck typing (lo que importa es que el objeto tenga el método, no su tipo).

El polimorfismo en métodos se ve cuando una subclase redefine (override) un método de la superclase.

El duck typing se resume en la idea:
_"Si se comporta como un pato, lo trato como un pato",_ es decir, lo importante es la interfaz de métodos, no la clase concreta.

El polimorfismo es clave para escribir código flexible, extensible y reutilizable, y Python lo potencia al no forzar tipos estáticos estrictos, sino enfocarse en lo que los objetos pueden hacer.

### Métodos y Operadores Especiales

En Python, los métodos y operadores especiales son funciones predefinidas que permiten personalizar el comportamiento de las clases y objetos. Estos métodos tienen nombres específicos que comienzan y terminan con dobles guiones bajos (`__`), y se utilizan para definir cómo los objetos de una clase interactúan con operadores y funciones integradas de Python.

### Métodos Dunder

Los métodos dunder (double underscore) son métodos especiales en Python que permiten definir el comportamiento de los objetos en diversas situaciones, como la creación, representación, comparación y operaciones aritméticas. Algunos ejemplos comunes de métodos dunder incluyen:

- `__init__(self, ...)`: Constructor que se llama cuando se crea una instancia de la clase.
- `__str__(self)`: Define la representación en cadena del objeto para la función `str()`.
- `__repr__(self)`: Define la representación oficial del objeto para la función `repr()`.
- `__add__(self, other)`: Define el comportamiento del operador de suma (`+`).
- `__sub__(self, other)`: Define el comportamiento del operador de resta (`-`).
- `__eq__(self, other)`: Define el comportamiento del operador de igualdad (`==`).

En fin, hay muchos más.

### Sobrecarga de Operadores

La sobrecarga de operadores en Python permite definir cómo los objetos de una clase interactúan con los operadores aritméticos y de comparación. Al implementar métodos dunder específicos, puedes personalizar el comportamiento de los operadores para tus objetos.
Ejemplo de Sobrecarga de Operadores en Python:

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
# Crear dos vectores
v1 = Vector(2, 3)
v2 = Vector(5, 7)

# Sumar los vectores utilizando el operador +
v3 = v1 + v2
print(v3)

# Restar los vectores utilizando el operador -
v4 = v1 - v2
print(v4)
````

En este ejemplo vemos que la clase `Vector` sobrecarga el operador de suma (`+`) mediante el método `__add__`, permitiendo sumar dos vectores de manera intuitiva.

### Cuando conviene usarlos?

Los métodos dunder y la sobrecarga de operadores son herramientas poderosas en Python que permiten personalizar el comportamiento de las clases y objetos. Sin embargo, es importante usarlos con criterio para evitar complicaciones innecesarias. Aquí hay algunas situaciones en las que conviene utilizarlos:

1. **Representación de Objetos**: Utiliza `__str__` y `__repr__` para proporcionar representaciones legibles y útiles de tus objetos. Esto es especialmente útil para depuración y registro.
2. **Operaciones Aritméticas**: Si tu clase representa una entidad matemática (como vectores, matrices, etc.), la sobrecarga de operadores como `__add__`, `__sub__`, `__mul__`, etc., puede hacer que el código sea más intuitivo y fácil de leer.
3. **Comparación de Objetos**: Implementa métodos como `__eq__`, `__lt__`, `__gt__`, etc., para definir cómo se comparan los objetos de tu clase. Esto es útil cuando necesitas ordenar o comparar instancias.
4. **Interacción con Funciones Integradas**: Si deseas que tus objetos funcionen bien con funciones integradas de Python (como `len()`, `iter()`, etc.), implementa los métodos dunder correspondientes (`__len__`, `__iter__`, etc.).
5. **Consistencia y Claridad**: Asegúrate de que la implementación de los métodos dunder sea consistente con las expectativas del usuario. Por ejemplo, si sobrecargas el operador de suma, asegúrate de que el comportamiento sea lógico y coherente con la naturaleza de los objetos que estás representando.

## Buenas prácticas en Programación Orientada a Objetos (OOP) 🧠

La Programación Orientada a Objetos no se trata solo de usar clases; su verdadero valor aparece cuando se diseña software **modular, reusable, extensible y fácil de mantener**.  
A continuación se presentan **buenas prácticas universales de OOP**, aplicables a cualquier lenguaje (incluyendo Python).

#### 1. Modelar objetos del mundo real con precisión
- Crear clases y atributos que representen conceptos **reales del dominio**.
- Si un objeto del mundo real **no existe y no aporta valor**, no debe convertirse en clase.
- Evitar clases “artefactuales” que no representan nada concreto.

Ejemplo:
En un e-commerce, es razonable modelar `Producto`, `Pedido`, `Cliente`.  
No es razonable modelar `AdministradorDeDiccionariosTemporales`.

#### 2. Una clase debe tener un único propósito
- Una clase bien diseñada responde a una **sola responsabilidad**.
- Si una clase hace varias cosas, debe dividirse.

Mal ejemplo:
`GestorPedidos` que además maneja pagos, facturas y correos.

Buen ejemplo:
`GestorPedidos`, `ProcesadorPagos`, `EmisorFacturas`, `EnviadorEmail`.


#### 3. Ocultar los detalles internos (Encapsulamiento)
- Los atributos deben ser privados o protegidos cuando corresponda.
- Publicar solo lo que es necesario para el uso de la clase.

Pautas:
- Exponer **métodos públicos** para interactuar con el objeto.
- Mantener **atributos y lógica interna encapsulados** para evitar manipulación externa indebida.


#### 4. Diseñar pensando en la extensión, no en la modificación
- Nuevos comportamientos deben agregarse **sin modificar código existente**.
- Usar herencia, polimorfismo, y especialmente **composición** para sumar capacidades.

Ejemplo:
Agregar un nuevo tipo de método de pago sin alterar las clases que ya funcionan.


#### 5. Usar Polimorfismo en lugar de condicionales
- Evitar `if-else` o `switch` basados en tipos de objetos.
- Definir métodos comunes en una interfaz o clase base, y dejar que las subclases implementen su comportamiento específico.

Mal Ejemplo:
```python
def procesar_pago(metodo_pago):
    if metodo_pago.tipo == "tarjeta":
        # lógica para tarjeta
    elif metodo_pago.tipo == "paypal":
        # lógica para PayPal
```

Buen Ejemplo:
```python
class MetodoPago:
    def procesar(self):
        pass
class Tarjeta(MetodoPago):
    def procesar(self):
        # lógica para tarjeta
class PayPal(MetodoPago):
    def procesar(self):
        # lógica para PayPal


def procesar_pago(metodo_pago: MetodoPago):
    metodo_pago.procesar()
```

#### 6. Evitar el "God Object"
- No crear clases que acumulen demasiadas responsabilidades o datos.
- Dividir en clases más pequeñas y especializadas.
- Un "God Object" es difícil de mantener y entender. Suele tener +20 métodos, +20 atributos.
- Dividir siempre!

#### 7. Nombrar Clases por lo que son, no por lo que hacen.
- Nombre de clase: sustantivo (ej.: Cliente, Factura, CarritoCompras)
- Nombre de método: verbo (ej.: calcular_total(), agregar_producto())
Esto hace el código más legible y profesional.

#### 8. Mantener métodos pequeños y específicos
- Un método debe hacer una sola cosa.
- Si un método supera 20–30 líneas, probablemente deba dividirse.

#### 9. No abusar de la herencia
Usar herencia cuando la relación es clara, directa y lógica.
Evitar:
   * Herencias excesivamente profundas (árboles de clases infinitos)
   * Herencia solo “para reutilizar código”
   * Heredar de una clase incompatible conceptualmente
Si se fuerza la herencia → usar composición.

#### 10. Escribir clases fáciles de testear
Una clase dependiente de otras clases debería:
   - Recibir sus dependencias como argumentos (inyección de dependencias)
   - No crear objetos “por dentro” constantemente
   - Permitir mocks en tests unitarios

#### 11. Documentar la intención, no solo el funcionamiento
El código debe responder, al menos con comentarios:
   - ¿Por qué esta clase existe?
   - ¿Cuál es su responsabilidad?
   - ¿Qué contrato promete cumplir?

#### 12. Seguir principios SOLID
Los principios SOLID son un conjunto de cinco principios de diseño orientado a objetos que ayudan a crear software más mantenible, flexible y escalable. Estos principios son:
- **S**ingle Responsibility Principle (SRP): Una clase debe tener una sola razón para cambiar.
- **O**pen/Closed Principle (OCP): Las clases deben estar abiertas para extensión, pero cerradas para modificación.
- **L**iskov Substitution Principle (LSP): Las subclases deben ser sustituibles por sus superclases.
- **I**nterface Segregation Principle (ISP): Muchas interfaces específicas son mejores que una interfaz general.
- **D**ependency Inversion Principle (DIP): Depender de abstracciones, no de concreciones.

Para más información, referirse a este [artículo](https://en.wikipedia.org/wiki/SOLID).
